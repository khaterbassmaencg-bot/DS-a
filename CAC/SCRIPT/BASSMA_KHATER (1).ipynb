## Plot Histograms for Numerical Features

### Subtask:
Generate histograms for the numerical features in `X` to visualize their distributions (e.g., 'age', 'trestbps', 'chol', 'thalach', 'oldpeak'). Ensure that missing values are handled appropriately for visualization, for example by dropping rows with NaN for the purpose of plotting or plotting only non-NaN values.


**Reasoning**:
I need to import the necessary plotting libraries, define the numerical features, and then iterate through them to generate histograms as per the instructions in the subtask. This will visualize the distributions of the specified numerical features.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the list of numerical features to plot
numerical_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

# Generate histograms for each numerical feature
for feature in numerical_features:
    plt.figure(figsize=(8, 5))
    sns.histplot(X[feature].dropna(), kde=True)
    plt.title(f'Distribution of {feature.replace("_", " ").title()}')
    plt.xlabel(feature.replace("_", " ").title())
    plt.ylabel('Frequency')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()
```

## Plot Count Plots for Categorical Features

### Subtask:
Generate count plots for the categorical and ordinal features in `X` to visualize their distributions (e.g., 'sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'). Handle missing values, if any, for these plots and ensure labels are clear.


**Reasoning**:
I need to define the list of categorical features, then iterate through them to generate count plots for each, handling missing values and setting appropriate titles and labels.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the list of categorical/ordinal features to plot
categorical_features = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']

# Generate count plots for each categorical/ordinal feature
for feature in categorical_features:
    plt.figure(figsize=(8, 5))
    sns.countplot(x=feature, data=X.dropna(subset=[feature]))
    plt.title(f'Distribution of {feature.replace("_", " ").title()}')
    plt.xlabel(feature.replace("_", " ").title())
    plt.ylabel('Count')
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.show()
```

**Reasoning**:
Now that the count plots for categorical features have been generated, the next step is to visualize the distribution of the target variable 'y' ('num'). A count plot is appropriate for this variable as it represents discrete categories.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a count plot for the target variable 'y'
plt.figure(figsize=(8, 5))
sns.countplot(x='num', data=y)
plt.title('Distribution of Target Variable (num)')
plt.xlabel('Diagnosis of Heart Disease (num)')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
